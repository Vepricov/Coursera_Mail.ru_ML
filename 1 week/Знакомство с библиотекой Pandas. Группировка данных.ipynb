# Знакомство с библиотекой Pandas. Группировка данных

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('bike_sharing_demand.csv', sep = ',')
```

```python
df.head(3)
```

```python
df.groupby(['casual'])
```

```python
df.groupby(['casual']).groups
```

```python
df.groupby(['casual']).first()
```

```python
df.groupby(['casual'])[['temp', 'atemp']].mean()
```

```python
df.groupby(['casual', 'weather'])[['temp']].mean()
```

```python
df.groupby(['casual']).agg({'atemp': sum, 'temp': 'first'})
```

```python
df.groupby(['casual']).agg({'atemp': [sum, 'mean'], 'temp': 'first'})
```

```python
df.groupby(['casual']).agg({'atemp': lambda x: max(x) + x.mean(), 'temp': 'first'})
```
